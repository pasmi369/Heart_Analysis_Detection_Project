```python
# convert age to years
cardio_df['age_year'] = (cardio_df['age']/365).round(0)
cardio_df.head()
```

```python
max(cardio_df['age_year'])
```

```python
min(cardio_df['age_year'])
```

```python
size_bins = [0,40,49,59,90]
group_names = ["<40","40-49", "50-59", "60-89"]
```

```python
cardio_df["age_group"] = pd.cut(cardio_df["age_year"], size_bins, labels = group_names)
cardio_df
```

```python
cardio_df.groupby("age_group").mean()["cholesterol"]
```

```python
cardio_df.groupby("age_group").mean()["gluc"]
```

```python
#converting height into meters
height_meters = cardio_df["height"]/100
cardio_df["bmi"] = round(cardio_df["weight"]/(height_meters**2),2)
cardio_df
```

```python
cardio_df = cardio_df[["age","gender","height","weight","bmi","ap_hi","ap_lo", "cholesterol","gluc","smoke","alco","active","cardio","age_year","age_group"]]
cardio_df
```
